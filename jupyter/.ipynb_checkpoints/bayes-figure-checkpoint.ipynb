```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from scipy import stats
import sqlite3
```

```python
directory = os.path.expanduser('~/Documents/data/')
con = sqlite3.connect(directory + 'cylinder-1/cylinder-1b.db')
```

```python
data = pd.read_sql("select * from measurement where grasp_prob_std > 0", con)
data['error'] = abs(data.grasp_prob - data.did_grasp)

plt.plot(data.did_grasp.rolling(window=100).mean())
plt.plot(data.error.rolling(window=100).mean())
plt.xlabel('Number')
plt.ylabel('Did Grasp');
```

```python


data = pd.read_sql("select * from measurement where grasp_method like 'BAYES_PROB' and grasp_prob_std > 0", con)
data['error'] = abs(data.grasp_prob - data.did_grasp)

slope, intercept, r_value, p_value, std_err = stats.linregress(np.sqrt(data.error), data.grasp_prob_std)

x = np.arange(0.0, 1.0, 0.01)
line = slope * np.sqrt(x) + intercept
plt.plot(x, line, 'r', label='y={:.2f}x+{:.2f}'.format(slope,intercept))

sigmoidi = lambda x: np.log(x) - np.log(1-x)
sigmoid = lambda x: 1/(1+np.exp(-x))
sigmoidd = lambda x: sigmoid(x) * (1 - sigmoid(x))

plt.scatter(data.error, data.grasp_prob_std, alpha=0.3)
plt.title('N={}'.format(len(data)))
plt.xlabel('Prediction Error')
plt.ylabel('Prediction Std');
```

```python
mid_bounds = []
upper_bounds = []
means = []
stds = []
diffs = []
probs = []
bins = 10
for i in range(bins):
    lower_bound = 1.0 / bins * i
    upper_bound = 1.0 / bins * (i + 1)
    mid_bound = (lower_bound + upper_bound) / 2
    data_bin = data[lower_bound < data.grasp_prob][data.grasp_prob < upper_bound]
    mid_bounds.append(mid_bound)
    upper_bounds.append(upper_bound)
    means.append(data_bin.did_grasp.mean())
    diffs.append(abs(data_bin.did_grasp.mean() - mid_bound))
    probs.append(data_bin.grasp_prob.mean())
    stds.append(data_bin.grasp_prob_std.mean())

plt.plot(mid_bounds, mid_bounds)
plt.plot(mid_bounds, means)
plt.plot(mid_bounds, np.array(stds) * np.abs(sigmoidi(np.array(mid_bounds))))
plt.plot(mid_bounds, diffs)
plt.title('N={}'.format(len(data)))
plt.xlabel('prob')
plt.ylabel('did_grasp mean');
```
